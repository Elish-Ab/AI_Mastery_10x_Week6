```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train_final = pd.read_csv('train_final.csv')
```

```python
train_final.head()
```

```python
# drop all nan rows

train_final.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split

# Assuming your dataset is `train_final`
X = train_final.drop('FraudResult_x', axis=1)
y = train_final['FraudResult_x']

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Initialize models
log_reg = LogisticRegression(max_iter=1000)
random_forest = RandomForestClassifier()
```

```python
# Train Logistic Regression
log_reg.fit(X_train, y_train)

# Train Random Forest
random_forest.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV

# Example for Logistic Regression
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}
grid_log_reg = GridSearchCV(log_reg, param_grid, cv=5)
grid_log_reg.fit(X_train, y_train)

# Example for Random Forest
param_grid_rf = {'n_estimators': [100, 200, 300], 'max_depth': [5, 10, 20]}
grid_rf = GridSearchCV(random_forest, param_grid_rf, cv=5)
grid_rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Predictions
y_pred_log_reg = log_reg.predict(X_test)
y_pred_rf = random_forest.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred_log_reg))

# Precision
print("Precision:", precision_score(y_test, y_pred_log_reg))

# Recall
print("Recall:", recall_score(y_test, y_pred_log_reg))

# F1 Score
print("F1 Score:", f1_score(y_test, y_pred_log_reg))

# ROC-AUC
print("ROC-AUC:", roc_auc_score(y_test, log_reg.predict_proba(X_test)[:, 1]))
```
